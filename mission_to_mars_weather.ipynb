```python
from bs4 import BeautifulSoup as bs
import requests
import pymongo
from pymongo import MongoClient
```

```python
url = 'https://twitter.com/marswxreport?lang=en'
response = requests.get(url)
```

```python
soup = bs(response.text, 'html.parser')
soup
```

```python
mars_weather = soup.find('div',class_= 'tweet').find('p')
weather_string = mars_weather.text
```

```python
data = []
weather_dict = {"Latest_Weather" : weather_string}
```

```python
data.append(weather_dict)
data
```

```python
conn = 'mongodb://localhost:27017'
client = pymongo.MongoClient(conn)
db = client.Mars_weather
for weather_data in data:
    db.weather.insert_one(weather_data)
```
